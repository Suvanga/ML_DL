```python
from sklearn.datasets import fetch_openml
import pandas as pd

# Load Boston Housing dataset from OpenML
df = fetch_openml(name="boston", version=1, as_frame=True)

# Convert to DataFrame
dataset = df.frame  # This automatically combines data + target
dataset.head()
```

```python
dataset['Price'] = df.target
dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
  # all rows, only the last column

X.head()
```

```python
y.head()
```

```python
#Linear Regression Model

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
X = dataset.drop(['Price', 'MEDV'], axis=1, errors='ignore')  # all features except target
y = dataset['Price']                                           # target

Lin_reg = LinearRegression()
X = X.astype(float)
y = y.astype(float)


mse = cross_val_score(Lin_reg, X, y, scoring='neg_mean_squared_error', cv=5)
meanmse = np.mean(mse)
print("Mean MSE:", meanmse)
```

```python
# now import train_test_split and then do the linear regression using train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
